```python

#EXERCISE #1

from collections import defaultdict, Counter


SEED_NUM = 1 
STUDENT_MAJOR = "BSECE"  
FAVORITE_ARTIST = "BEN&BEN"
LOCAL_HAZARD = "EARTHQUAKE"
CONTROL_NUM = max(3, SEED_NUM)


stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]


expansion_count = CONTROL_NUM + 3
custom_tuples = [("TUP Manila", "BSIT"), ("TUP Taguig", "BSIT"), ("TUP Manila", "BSECE")] * expansion_count

stream.extend(custom_tuples[:expansion_count])


campus_groups = defaultdict(list)
for campus, program in stream:
    campus_groups[campus].append(program)


campus_counts = {}
for campus, programs in campus_groups.items():
    campus_counts[campus] = Counter(programs)


manila_top = campus_counts["TUP Manila"].most_common(1)

print(f"{'CAMPUS':<15} | {'TOP PROGRAM':<15} | {'FREQUENCY'}")
print("-" * 45)
for campus, counts in campus_counts.items():
    top_prog, freq = counts.most_common(1)[0]
    print(f"{campus:<15} | {top_prog:<15} | {freq}")

print("\nSpecific Requirement:")
if manila_top:
    print(f"The top requested program in Manila is {manila_top[0][0]} with {manila_top[0][1]} requests.")
```

```python
#EXERCISE #2

CONTROL_NUM = 3
FAVORITE_ARTIST = "BEN&BEN" 


festival = {"BEN&BEN", "SB19", "Bini", "Eraserheads", FAVORITE_ARTIST, "Zild", f"Indie Artist {CONTROL_NUM}"}

user_a = {"BEN&BEN", "Bini", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"}


new_artists = {f"New Artist {i}" for i in range(1, CONTROL_NUM + 1)}
festival.update(new_artists)




intersection_a = user_a.intersection(festival)
union_a = user_a.union(festival)
similarity_a = (len(intersection_a) / len(union_a)) * 100


intersection_b = user_b.intersection(festival)
union_b = user_b.union(festival)
similarity_b = (len(intersection_b) / len(union_b)) * 100


dealbreakers_a = user_a.difference(festival)


print(f"Updated Festival Lineup Size: {len(festival)}")
print(f"User A Similarity: {similarity_a:.2f}%")
print(f"User B Similarity: {similarity_b:.2f}%")
print(f"User A Dealbreakers: {dealbreakers_a}")
```

```python
# EXERCISE 3

from collections import deque, Counter


CONTROL_NUM = 3
LOCAL_HAZARD = "EARTHQUAKE"


rolling_buffer = deque(maxlen=5)

burst = [
    (CONTROL_NUM, "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, LOCAL_HAZARD),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]


last_timestamp = burst[-1][0]
expansion_count = CONTROL_NUM + 2

for i in range(1, expansion_count + 1):
   
    category = "TRAFFIC" if i % 2 == 0 else "FIRE"
    burst.append((last_timestamp + i, category))


rejected_count = 0

for alert in burst:
    
    current_categories = [item[1] for item in rolling_buffer]
    
    
    counts = Counter(current_categories)
    
   
    incoming_category = alert[1]
    if counts[incoming_category] >= 2:
        rejected_count += 1
        
        continue
    else:
        rolling_buffer.append(alert)


total_timestamp_sum = sum(alert[0] for alert in rolling_buffer)


print(f"Total Alerts in Expanded Burst: {len(burst)}")
print(f"Final Deque State: {list(rolling_buffer)}")
print(f"Number of Alerts Rejected as Spam: {rejected_count}")
print(f"Sum of Stored Timestamps: {total_timestamp_sum}")
```
