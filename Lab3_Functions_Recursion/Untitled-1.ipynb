```python
#CELL 1

LAST_NAME = "Araos"
STUDENT_ID = "TUPM-25-3591"

SEED_DIGIT = int(STUDENT_ID[-1])
ID_SUM = SUM = sum (int(d)for d in STUDENT_ID if d.isdigit())
NAME_LENGTH = len(LAST_NAME)

print("Student:", LAST_NAME)
print("Seed Digit:", ID_SUM)
print("Surname Length:", NAME_LENGTH)
```

```python
#CELL 4

def user_summary(title, *scores, **info):
    print(f"==={title} ===")

    total = sum(scores)
    average = total / len(scores)

    print("Scores:", scores)
    print("Total:", total)
    print("Average:", round(average, 2))

    for k,v in info.items():
        print(f"{k}: {v}")

        return average

avg = user_summary(
    f"{LAST_NAME} Academic Report",
    SEED_DIGIT * 10,
    ID_SUM % 100,
    NAME_LENGTH * 5, 
    id=STUDENT_ID,
    surname=LAST_NAME
)
```

```python
# CELL 5

def compute_area(radius):
    return 3.1416 * radius**2

area = compute_area(SEED_DIGIT + 2)
print("Computed Area:", round(area,2))
```

```python
#CELL 6

def safe_calculator(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        result = "UNDEFINED"
    return result

print(ID_SUM)
print(SEED_DIGIT)
print("Calculation Result:", safe_calculator(ID_SUM, SEED_DIGIT))
```

```python
#CELL 7

def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)

print(SEED_DIGIT)
print("Factorial Result", factorial(SEED_DIGIT))
```

```python
#CELL 8

def reverse(text):
    if len(text) <=1:
        return text
    return text [-1] + reverse(text[:-1])

print("Reversed Surname:", reverse(LAST_NAME.upper()))
```

```python
#CELL 9

def logger(func):
    def wrapper(*args, **kwargs):
        print("Executing:", func.__name__)
        return func(*args, **kwargs)
    return wrapper

@logger
def multiply(x,y):
    return x * y

print ("Multiplication:", multiply(SEED_DIGIT, NAME_LENGTH))
```

```python
# CELL 10

data = list (range(1, SEED_DIGIT + 10))
squared = list(map(lambda x: x**2, data))

print(data)
print("Squared Data", squared)
```

```python
# CELL 11

def even_squares(limit):
    for i in range(limit):
        if i % 2 == 0:
            yield i**2
       
       
for value in even_squares(NAME_LENGTH + SEED_DIGIT):
    print(value)
```
